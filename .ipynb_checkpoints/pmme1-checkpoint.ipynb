### Predictive Maintenence for Military Equipment

```python
# import all necessary modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Setup display options
pd.set_option('display.max_columns', 30)
```

```python
# Load the dataset
df = pd.read_csv('train_FD001.txt', sep=r'\s+', header=None)

# Show first 5 rows
df.head()
```

```python
# Define column names for the 26 columns (1 engine_id + 1 cycle + 3 settings + 21 sensors)
column_names = ['engine_id', 'cycle'] + \
               [f'op_setting_{i}' for i in range(1, 4)] + \
               [f'sensor_{i}' for i in range(1, 22)]

df.columns = column_names

# Confirm structure
df.head()
```

```python
df.describe()
```

```python
# Get the max cycle for each engine
max_cycle = df.groupby('engine_id')['cycle'].max().reset_index()
max_cycle.columns = ['engine_id', 'max_cycle']

# Merge with original dataframe
df = df.merge(max_cycle, on='engine_id')
```
